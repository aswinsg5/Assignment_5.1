```python
import pandas as pd
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
baby_name = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv",sep = ',')
```

<b>Questions:</b>

<b>1. Delete unnamed columns</b>

```python
baby_name.drop(baby_name.columns[baby_name.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
baby_name.head(3)
```

<b>2. Show the distribution of male and female</b>

```python
new_gender = pd.Categorical(baby_name['Gender'])
new_gender.describe()
```

<b>3. Show the top 5 most preferred names</b>

```python
baby_name_count = baby_name[["Name","Count"]]
baby_name_count.groupby('Name').sum().sort_values(by = 'Count', ascending = False).head(5)
```

<b>4. What is the median name occurence in the dataset</b>

```python
baby_name_count[baby_name_count['Count'] == baby_name_count['Count'].median()]
```

<b>5. Distribution of male and female born count by states</b>

```python
byNameState = baby_name[['State','Gender','Count']]
byNameState.groupby(['State','Gender']).sum()
```
